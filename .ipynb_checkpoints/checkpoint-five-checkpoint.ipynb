```python
pd.set_option('display.float_format', lambda x: '%.2f' % x) #gets rid of scientific notation
df_covid_data[['Entity', 'Total confirmed deaths due to COVID-19']].groupby(['Entity']).describe()
```

```python
df_covid_data['Entity'].unique() #identify unique values so that I can remove continents and regions and only look at country data
```

```python
#make a list of entries to exclude
countries_exclude_list = ['Asia', 'Asia excl. China', 'Europe', 'European Union', 'High income', 'Low income', 'Lower middle income', 'North America', 'Oceania', 'South America', 'Upper middle income', 'World', 'World excl. China', 'World excl. China and South Korea', 'World excl. China, South Korea, Japan and Singapore']
# remove exclude list
#boolean mask for rows to keep (True) and remove (False)
mask = ~df_covid_data['Entity'].isin(countries_exclude_list)
#pply mask
filtered_df_covid_data = df_covid_data[mask]
filtered_df_covid_data['Entity'].unique()
```

```python
# only view these two variables
filtered_df_covid_data[['Entity', 'Total confirmed deaths due to COVID-19']].corr()
```

```python
fig, ax = plt.subplots(figsize=(14,8))

ax.set_ylabel('Total confirmed deaths due to COVID-19')

# boxplot with only these two variables
_ = df_covid_data[['Entity', 'Total confirmed deaths due to COVID-19']].boxplot(by='Entity', figsize=(13,8), vert=False, sym='b.', ax=ax)
```

```python
#select data for United states only in order to simplify code to start
countries_include_list = ['United States']

#boolean mask for rows to keep (True) and remove (False)
mask = filtered_df_covid_data['Entity'].isin(countries_include_list)
#pply mask
US_covid_data = filtered_df_covid_data[mask]
US_covid_data['Entity'].unique()
```

```python
#sort by day
US_covid_data = US_covid_data.sort_values('Day')
```

```python
#calculate daily new cases by subtractng from previous day count
US_covid_data['Daily new cases'] = US_covid_data['Total confirmed cases of COVID-19'] - US_covid_data['Total confirmed cases of COVID-19'].shift(1)
US_covid_data.head()
```

```python
US_covid_data['DayOfWeek'] = US_covid_data['Day'].dt.day_name()
```

```python
fig, ax = plt.subplots(figsize=(14,8))

ax.set_ylabel('Day of the week')

# boxplot
_ = US_covid_data[['DayOfWeek', 'Daily new cases']].boxplot(by='DayOfWeek', figsize=(13,8), vert=False, sym='b.', ax=ax)

ax.set_xlabel('Number of daily new cases')
```

```python
#order by day of week
US_covid_data['DayOfWeek'].unique()
```

```python
# fewer people are getting cases on Sunday, or fewer people are getting tested on Sunday - tell us about the disease or how people are collecting/reporting data
```

```python
# Trying another bar chart to show percentage deaths by county

filtered_df_covid_data['Percentage Deaths'] = (filtered_df_covid_data['Total confirmed deaths due to COVID-19'] / filtered_df_covid_data['Total confirmed cases of COVID-19']) * 100
#group by country/"Entity"
country_percentage_deaths = filtered_df_covid_data.groupby('Entity')['Percentage Deaths'].mean()

filtered_df_covid_data_sorted = filtered_df_covid_data.sort_values(by='Percentage Deaths', ascending=False)

plt.figure(figsize=(12, 6))  # Adjust figure size as needed
plt.bar(filtered_df_covid_data_sorted['Entity'], filtered_df_covid_data_sorted['Percentage Deaths'])
plt.xlabel('Entity')
plt.ylabel('Percentage of Deaths')
plt.title('Percentage of Deaths due to Covid-19 by County')
ax.set_xlabel('x-axis', fontsize = 8) 
plt.xticks(rotation=90)  # Rotate x-axis to read
plt.tight_layout()  # Labels displayed
plt.show()
```
